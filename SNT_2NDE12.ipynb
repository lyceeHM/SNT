```python
import pandas as pa

T = pa.read_excel('/Users/fredericbro/Documents/Lycee(henrimoissan)/Prof (henri-moissan)/Annee 2020/Lycée/fichier_eleves.xlsx')
T.head()
```

```python
T.columns = T.columns.str.replace(' ', '_')
T['Division'] = T['Division'].str.replace(' ', '_', regex = True)
T.rename(columns = {'Nom_de_famille' : 'Nom', 'Prénom_1' : 'Prenom'}, inplace = True)
T.head()
```

```python
S12 = T.query('Division in "2NDE_12"')
S12 = S12[[c for c in T.columns if c not in  ["Option_5" ,"Option_6" ,"Option_7", "Option_8"]]]
S12.index = range(len(S12))
S12.head()
```

```python
S12.Nom = S12.Nom.str.lower()
S12.Prenom = S12.Prenom.str.lower()
S12.Prenom = S12.Prenom.str.replace('ë','e')
```

```python
S12['Num'] = list(range(len(S12)))

def mail(x):
    return x['Nom'][0]+'.'+x['Prenom']+str(x.Num)+'77@gmail.com'
S12['Mail'] = S12.apply(mail, axis = 1)
S12
```

```python
S12['Groupe'] = ""
S12['Distance'] = ""
S12.to_excel('2nde_12.xlsx', encoding = 'utf8', index = None)
```
